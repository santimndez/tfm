```python
import cv2 as cv
import numpy as np
import sys
from pathlib import Path

# Ensure the project's src directory is on sys.path so the local 'utils' package can be imported.
# This notebook lives in .../src/notebooks, so parent is .../src which contains the utils package.
sys.path.append(str(Path.cwd().parent))

from utils.ball_detection import get_ball_positions_from_masks, save_ball_positions

masks_folder = ['../../../videos/results/segmentation/sam3/masks_s1A',
                '../../../videos/results/segmentation/sam3/masks_s1b']
videos = ['../../../videos/test2/s/test2A_s1.mp4', 
          '../../../videos/test2/s/test2B_s1.mp4']
names = ['../data/positions_sam3_s1A.csv', '../data/positions_sam3_s1B.csv']
w, h = 1920, 1080 # Tama√±o de los frames de los videos

for name, video, folder in zip(names, videos, masks_folder):
    nframes = int(cv.VideoCapture(video).get(cv.CAP_PROP_FRAME_COUNT))
    positions = get_ball_positions_from_masks(folder, w, h, nframes)
    save_ball_positions(name, positions)
```
